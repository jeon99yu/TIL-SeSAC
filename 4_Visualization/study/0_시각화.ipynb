### 한번에 여러 그래프 그리기

```python
x = np.arange(-4.5, 5, 0.5)

y1 = x**2
y2 = x+30
y3 = x**2+10

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

### plot 옵션

```python
x = np.arange(-4.5, 5, 0.5)

y1 = x**2
y2 = x+30
y3 = x**2+10

plt.plot(x, y1, color="red", linestyle="dashed")
plt.plot(x, y2, "b:")
plt.plot(x, y3, color='#00FF00', linestyle="-.")

plt.show()
```

### 축의 범위조정

```python
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y, '--r')
plt.axis([-20, 30, -20, 120])
#plt.xlim(-20, 30) #이 방법을 더 많이 쓰긴 함
#plt.ylim(-20, 120)
plt.show()
```

### label 과 legend

```python
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = x ** 3

plt.plot(x, y1, 'r', label='라인1')
plt.plot(x, y2, 'g', label='라인2')

plt.xlabel('x 라벨')
plt.ylabel('y 라벨')

plt.legend(loc=0) # 이런식으로 더 많이 씀 loc = 0이 보통 best위치
# plt.legend(bbox_to_anchor=(1.04, 1)) # 정확한 포인트 위치조정
plt.show() 
```

### subplot

```python
plt.subplot(131) # 1행 3열의 1번째 위치 (행크기, 열크기, 순서)
plt.plot([1, 2], color='r')

plt.subplot(132)
plt.plot([1, 2], color='g')

plt.subplot(133)
plt.plot([1, 2], color='b')

plt.show()
```

```python
plt.subplot(221)
plt.plot([1, 2], color='r')

plt.subplot(222)
plt.plot([1, 2], color='g')

plt.subplot(223)
plt.plot([1, 2], color='b')

plt.subplot(224)
plt.plot([1, 2], color='r')

plt.show()
```

### Grid

```python
# figsize = 캔버스 크기 설정
plt.figure(figsize=(5, 5))

grid = plt.GridSpec(2, 3)

plt.subplot(grid[0, 0])
plt.plot([1, 2], color='r')

plt.subplot(grid[0, 1])
plt.plot([1, 2], color='g')

plt.subplot(grid[0, 2])
plt.plot([1, 2], color='b')

plt.subplot(grid[1, :2])
plt.plot([1, 2], color='b')

plt.subplot(grid[1, 2])
plt.plot([1, 2], color='r')

plt.show()
```

### Scatter (산점도)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(30)
y = np.random.rand(30)
a = np.random.rand(30)
b = np.random.rand(30)
e = np.random.rand(30)
f = np.random.rand(30)
color = np.random.rand(30)

size = 100 * np.random.rand(30)

plt.scatter(x, y, s=size, c=color, alpha=0.5, marker='o')
plt.scatter(a, b, s=size, c=color, alpha=0.5, marker='^')
plt.scatter(e, f, s=size, c=color, alpha=0.5, marker='+')
plt.show()
```

### Histogram (히스토그램)

```python
data = [np.random.randint(1, 7) for i in range(10)]
print(data)
plt.hist(data, bins=6, alpha=1)
plt.show()
```

### Bar (바차트)

```python
# 세로 그래프 -> bar / 가로 그래프 ->  barh(horizon)
plt.bar(
    ['10대', '20대', '30대', '40대', '50대'], # x축 카테고리형
    [np.random.randint(1, 50) for i in range(5)]) # y축 연속형
#plt.barh(['10대', '20대', '30대', '40대', '50대'], [np.random.randint(1, 50) for i in range(5)], color='g')
plt.title('제목')
plt.xlabel('나이')
plt.ylabel('값')
plt.show()
```
