### 할리스커피 크롤링

- 매장정보 크롤링 + 총 10페이지

```python
import requests
from bs4 import BeautifulSoup

stores = []

for page in range(1,11):
    res = requests.get('https://www.hollys.co.kr/store/korea/korStore2.do')
    soup = BeautifulSoup(res.text, 'html.parser')
    tag_tbody = soup.find('tbody')
    
    for store in tag_tbody.find_all('tr'):
        store_td = store.find_all('td')
        if len(store_td) < 6:
            continue

        # 매장 서비스: <img alt="테라스"> <img alt="주차"> 형태로 들어있음
        service_imgs = store_td[4].find_all('img')
        services = [img['alt'] for img in service_imgs]  # alt 값 리스트로 추출
        service_text = ', '.join(services)  # 문자열로 연결

        stores.append({
            '지역': store_td[0].text.strip(),
            '매장명': store_td[1].text.strip(),
            '현황': store_td[2].text.strip(),
            '주소': store_td[3].text.strip(),
            '매장 서비스': service_text,
            '전화번호': store_td[5].text.strip()
        })
    
stores
```

```python
import requests
from bs4 import BeautifulSoup

stores = []

for page in range(1, 11):
    params = {'pageNo': page}
    res = requests.post('https://www.hollys.co.kr/store/korea/korStore2.do', data=params)
    soup = BeautifulSoup(res.text, 'html.parser')

    tag_tbody = soup.find('tbody')

    for store in tag_tbody.find_all('tr'):
        store_td = store.find_all('td')
        if len(store_td) < 6:
            continue

        store_info = {
            '지역': store_td[0].text.strip(),
            '매장명': store_td[1].text.strip(),
            '현황': store_td[2].text.strip(),
            '주소': store_td[3].text.strip(),
            '매장 서비스': store_td[4].text.strip(),
            '전화번호': store_td[5].text.strip()
        }
        stores.append(store_info)

for store in stores:
    print(store)
```
