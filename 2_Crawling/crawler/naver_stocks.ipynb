### 네이버 주가 크롤링

1. 종목명과 현재가를 크롤링해주세요

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://finance.naver.com/sise/sise_quant.nhn')
soup = BeautifulSoup(res.text, 'html.parser')

rows = soup.select('.type_2 tr')
stocks = []

for row in rows:
    name_tag = row.select_one('a.tltle') # 종목명
    price_tag = row.select_one('td.number') # 현재가

    if name_tag and price_tag:
        stock = {
            'name': name_tag.text.strip(),
            'price': price_tag.text.strip()
        }
        stocks.append(stock)

stocks
```

2. 위 종목 정보에서, 전일 대비 상승한 종목만 수집해주세요

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://finance.naver.com/sise/sise_quant.nhn')
soup = BeautifulSoup(res.text, 'html.parser')

rows = soup.select('.type_2 tr')
rising_stocks = []

for row in rows:
    name_tag = row.select_one('a.tltle') 
    price_tag = row.select_one('td.number') 
    change_tag = row.select_one('span') 

    if name_tag and price_tag and change_tag and change_tag.text.strip() == '상승':
        stock = {
            'name': name_tag.text.strip(),
            'price': price_tag.text.strip()
        }
        rising_stocks.append(stock)
        
rising_stocks
```
